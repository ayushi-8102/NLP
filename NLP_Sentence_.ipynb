```python
import pandas as pd
import numpy as np
np.random.seed(2022)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
data=messages = pd.read_csv('software_requirements_extended.csv',
                           names=["label", "requirement"])
```

```python
data.head()
```

```python
#Inpedendent Features
z=list(data['requirement'])
y=list(data['label'])
```

```python
y
```

```python
import spacy
import string
nlp = spacy.load("en_core_web_sm")
stop_words = nlp.Defaults.stop_words
print(stop_words)
```

```python
punctuations = string.punctuation
print(punctuations)
```

```python
# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    doc = nlp(sentence)



    # print(doc)
    # print(type(doc))

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() for word in doc ]

    # print(mytokens)

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]
    
    sentence = " ".join(mytokens)
    # return preprocessed list of tokens
    return sentence
```

```python
spacy_tokenizer("The system shall")
```

```python
data['tokenize'] = data['requirement'].apply(spacy_tokenizer)
```

```python
data.head()
```

```python
data['embeddings'] = data['tokenize'].apply(model.encode)
```

```python
data.head()
```

```python
X = data['embeddings'].to_list()
y=list(data['label'])
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
y=LabelEncoder().fit_transform(y)
```

```python
y
```

```python
### Train Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
from sklearn import metrics
predicted = LR.predict(X_test)

print("Logistic Regression Accuracy:",metrics.accuracy_score(y_test, predicted))
print("Logistic Regression Precision (Macro):", metrics.precision_score(y_test, predicted, average='macro'))
print("Logistic Regression Recall (Macro):", metrics.recall_score(y_test, predicted, average='macro'))
```
